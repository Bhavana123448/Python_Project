```python
# -------------------------------
# Python Number Guessing Game
# Colab-friendly version
# -------------------------------

import random
from IPython.display import clear_output  # For Colab-friendly screen clearing

def play_game():
    clear_output()  # Clears previous output for a clean game interface
    print("🎮 Welcome to the Number Guessing Game!\n")

    # Difficulty selection
    print("Select Difficulty Level")
    print("1. Easy (10 Attempts)")
    print("2. Medium (7 Attempts)")
    print("3. Hard (5 Attempts)")

    choice = input("Enter choice (1/2/3): ")

    if choice == '1':
        attempts = 10
    elif choice == '2':
        attempts = 7
    elif choice == '3':
        attempts = 5
    else:
        print("Invalid choice. Default difficulty set to Medium.")
        attempts = 7

    number = random.randint(1, 100)
    print(f"\nI have selected a number between 1 and 100. You have {attempts} attempts to guess it.\n")

    # Game loop
    for i in range(attempts):
        try:
            guess = int(input(f"Attempt {i+1}: Enter your guess: "))
            if guess > number:
                print("Too High!\n")
            elif guess < number:
                print("Too Low!\n")
            else:
                print(f"🎉 Congratulations! You guessed the number in {i+1} attempts!")
                break
        except ValueError:
            print("⚠ Please enter a valid number.\n")
    else:
        print(f"❌ Game Over! The number was: {number}")

# -------------------------------
# Replay loop
# -------------------------------
while True:
    play_game()
    again = input("\nDo you want to play again? (yes/no): ").lower()
    if again != "yes":
        print("Thanks for playing! 👋")
        break
```
